```python
#we can define a function to do everything at once, which takes a file path without the extension
def modify_list(file):
    file_path = file + '.csv'
    queen = pd.read_csv(file_path)
    queen_modified = queen.drop_duplicates(subset=['Description', 'Price', 'Rating', 'ReviewCount'])
    queen_modified = queen_modified.sort_values(by='Description')
    queen_modified = queen_modified.dropna(axis = "rows")
    queen_modified_p = queen_modified.sort_values(by=['Rating', 'ReviewCount'], ascending=[False, False])
    #saving the files:
    queen_modified.to_csv('./modified/' + file + 'f.csv', index=False)
    queen_modified_p.to_csv('./optimal/' + file + 'p.csv', index=False)
    
```

```python
#testing the function, with the may 20 dataset
modify_list("Queen_mattress_may20_2022")

#it works, as it produces the output
```
